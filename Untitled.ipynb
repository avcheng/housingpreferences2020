```python
import csv
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('housing_prefs.csv')
del df['Please ONLY fill out this survey if you are a current freshman.']
df = df.drop(df.columns[[0, 13, 14, 15, 16, 17]], axis=1)
df.head()
```

```python
rankings = []
# get only the house columns
for column in df:
    rankings.append(df[column].mean())
    
rankings
```

```python
houses = ['Adams', 'Cabot', 'Currier', 'Dunster', 'Eliot', 'Kirkland', 
              'Leverett', 'Lowell', 'Mather', 'Pfoho', 'Quincy', 'Winthrop'] 

plot_df = pd.DataFrame({'x' : houses , 'y' : rankings})

plot_df = plot_df.sort_values('y')

plt.figure(figsize=(20, 8))
fig, ax = plt.subplots()
fig.set_size_inches(18.5, 10.5)

plot_df.plot(kind='bar', x='x', y='y', ax=ax, width=0.8)
ax.set_xlabel('Houses', fontsize = '20')
ax.set_ylabel('Average Ranking', fontsize = '20')
ax.get_legend().remove()

rects = ax.patches

rounded = [round(num, 3) for num in plot_df.y]


for rect, label in zip(rects, rounded):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 0.1, label,
            ha='center', va='bottom', fontsize='10')

plt.show()
```
